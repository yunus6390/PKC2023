```python
#  Assignment: how to change the x and y axis titles
```

```python
g= sns.FacetGrid(kashti,hue="survived",col="pclass",margin_titles=True,
              palette={1:"seagreen",0:"grey"})
g=g.map(plt.scatter,"fare","age",edgecolor= "w").add_legend();
```

```python
g= sns.FacetGrid(kashti,hue="survived",col="sex",margin_titles=True,
              palette={1:"seagreen",0:"grey"})
g=g.map(plt.scatter,"fare","age",edgecolor= "w").add_legend();
```

```python
g= sns.FacetGrid(kashti,hue="survived",col="sex",margin_titles=True,
              palette={1:"seagreen",0:"grey"})
g=g.map(plt.scatter,"fare","age",edgecolor= "w").add_legend();
```

```python
# heat map
corr = kashti.corr()
plt.figure(figsize=(10,10))
sns.heatmap(corr)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Exclude non-numeric columns
numeric_columns = kashti.select_dtypes(include=['number'])
corr = numeric_columns.corr()

# Plot the heatmap
plt.figure(figsize=(10, 10))
sns.heatmap(corr)
plt.title('Correlation Heatmap')
plt.show()
```
