# exploratory data analysis:
exploratory Data Analysis is a method used to `analyze` and `summarize` datasets. In python , we can use the pandas library for this purpose,

```python
df.drop(columns= "deck", inplace = True)
```

```python
df.shape
```

```python
# Visualize and explore
sns.countplot(data= df, x= 'survived')
```

```python
sns.barplot( data= df, x="survived", y="fare", hue= "who")
```

```python
sns.countplot(x="pclass", data=df)
```

```python
# data distribution
sns.histplot(data= df, x= "age", bins= 30)
```

```python
df['age'].hist(bins= 30, color = 'darkred', alpha= 0.75)
```

```python
sns.boxplot(data= df, y='age', x = 'survived', hue='who')
```

```python
# Scatterplot
sns.scatterplot(x= 'age', y='fare', data = df, hue='pclass')
plt.ylim(0,300)
```

```python
sns.pairplot(df, hue='survived')
```

```python
# stacked barchart/plot
df.groupby(['pclass', 'sex'])['survived'].mean().unstack().plot(kind= 'bar', stacked = True)
```

```python
df.groupby(['pclass', 'who'])['fare'].mean().unstack().plot(kind= 'bar', stacked = True)
```

```python
# swarmplot
sns.swarmplot(data=df, x="survived", y="age",hue="who")
```

```python
sns.catplot(x='pclass',y='fare', hue='embarked',kind= 'point',data= df)
```

```python
sns.lmplot(data= df[df['fare']<300], x= 'age', y= 'fare')
```

```python
sns.lmplot(data=df,x='age',y='fare', hue= 'pclass')
plt.ylim(0,300)
```

```python
# relation of two variables

import seaborn as sns

# Filter the DataFrame to include only fare values less than 300
filtered_df = df[df['fare'] < 300]

# Plot using lmplot
sns.lmplot(data=filtered_df, x='age', y='fare')
```
