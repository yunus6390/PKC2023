# 16- splitting a string into multiple columns

```python
import pandas as pd
df = pd.DataFrame({'name': ['Ahmad Raza','Ali Afzal','Sajjad Ali','Abu Bakar'],
                   'location': ['Pune,India','Mumbai,India','Delhi,India','Nashik,India']})
df
```

```python
# splitting a column into 2 columns
df.name.str.split(' ', expand=True)
```

```python
# adding those splits into new columns
df[["first_name","last_name"]] = df.name.str.split(' ', expand=True)
df
```

```python
df[["city","country"]] = df.location.str.split(',', expand=True)
df
```

```python
# Refine data manipulation
df = df[['first_name', 'last_name','city','country']]
df
```

```python
fasla.head().T
```
